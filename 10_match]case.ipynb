```python
color = input("Enter your color ")
match color:
    case "red":
        print("Input color is red")
    case "blue":
        print("color is blue")
    case "orange":
        print("Color is Orange")
    case _:
        print("Non of the color is matching")
```

```python
number = int(input("Enter your color "))
match number:
    case number if number<0:
        print("The numbrf is negative")
    case number if number==0:
        print("The number is zero")
    case _:
        print("Number is positive")
        
```

```python
day = int(input("Enter your number between 1-7"))
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
```

```python
month = input("Enert the month")
match month:
    case "December" | "Jan" | "Feb":
        print("Winter")
    case "mar"|"april"|"may":
        print("Summer")
    case "June"|"july"|"August":
        
        print("Rain")
    case _:
        print("Not a valid month")
```

```python
filename = input("Enter your file with extension").split(".")[1]
match filename:
    case 'jpg' |'png'|'gif':
        print("Your file type is image file")
    case 'py'|'ipynb'|'cpp':
        print("Your file type is coding")

    case 'pdf'|'docx'|'xlsx':
        print("Your file type is working file")
    case _:
        print("Not able to match")
```
